1. Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)

```python
import pandas as pd
```

```python
from contextlib import nullcontext
import seaborn as sns
penguins = sns.load_dataset("penguins")
```

```python
penguins
```

```python
penguins["beak_length_type"] = ''
penguins
```

```python
def break_type(size):
  if size > 42:
    return 'High'
  elif 35 <= size <= 42:
    return 'Medium'
  else:
    return 'Low'

penguins['beak_length_type'] = penguins['bill_length_mm'].apply(break_type)
penguins
```
